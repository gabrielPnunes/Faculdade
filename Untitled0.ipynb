### Questão 02

de nome demonstrar como retirar apenas o primeiro nome

```python
monta = ''
for let in nome:
    if let != ' ':
        monta += let
    else:
        break
print(monta)
```

###Questão 06

calcular quantos dias faltam para seu aniversário

```python
from datetime import date
data = date.fromisoformat('2005-07-03')
dataaniversario = date(2024,7,3)
aniversario = dataaniversario - data
print(aniversario)
```
